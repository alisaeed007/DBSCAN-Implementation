```python
# Importing required libraries
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Data points
data = np.array([
    [14, 32], [15, 32], [16, 32], [17, 32],
    [44, 66], [33, 22], [12, 67], [13, 43],
    [45, 23], [32, 32], [34, 43], [15, 31]
])

# DBSCAN Parameters
epsilon = 10  # Maximum distance between two samples for one to be considered as in the neighborhood of the other
min_samples = 2  # Minimum number of points to form a dense region

# Applying DBSCAN
clustering = DBSCAN(eps=epsilon, min_samples=min_samples)
clusters = clustering.fit_predict(data)

unique_clusters = set(clusters)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_clusters))]

for k, col in zip(unique_clusters, colors):
    if k == -1:  # Noise points
        col = [0, 0, 0, 1]  # Black color for noise

    class_member_mask = (clusters == k)

    xy = data[class_member_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        'o',
        markerfacecolor=tuple(col),
        markeredgecolor='k',
        markersize=14
    )

plt.title(f"DBSCAN Clustering (epsilon={epsilon}, min_samples={min_samples})")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```
